```python
import numpy as np
import pandas as pd

train_image_dir = '/content/drive/MyDrive/DL/images-final/images/train'
valid_image_dir = '/content/drive/MyDrive/DL/images-final/images/validate'
test_image_dir = '/content/drive/MyDrive/DL/images-final/images/test'
```

```python
import os
import pandas as pd

def create_dataframe_from_directory(directory_path):
    file_data = {'FILENAME': [], 'IDENTITY': []}

    for root, dirs, files in os.walk(directory_path):
        for file_name in files:
            full_path = os.path.join(root, file_name)
            relative_path = full_path.replace(directory_path, '')  # Replace the directory path with an empty string
            relative_path = relative_path.lstrip(os.path.sep)  # Remove leading separator if present
            file_data['FILENAME'].append(relative_path)
            file_data['IDENTITY'].append(file_name)

    dataframe = pd.DataFrame(file_data)
    return dataframe

# Create DataFrame from the directory
file_dataframe = create_dataframe_from_directory(train_image_dir)

# Display the created DataFrame
print(file_dataframe.head())
```

```python
file_dataframe['IDENTITY'] = [element[:-4] for element in file_dataframe['IDENTITY']]
```
